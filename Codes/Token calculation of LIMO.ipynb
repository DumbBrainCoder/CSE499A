```python
import pandas as pd
from transformers import AutoTokenizer

# 1. Load tokenizer (same as your model)
model_name = "Qwen/Qwen2.5-0.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 2. Load your Excel file
df = pd.read_excel("/content/output (7).xlsx")  # adjust path if needed

# 3. Make sure required columns exist
required = {"question", "solution", "answer"}
assert required.issubset(df.columns), f"Excel must contain {required}"

# 4. Function to build prompt and count tokens
def count_prompt_tokens(row):
    prompt = f"Question: {row['question']}\nReasoning: {row['solution']}\nAnswer: {row['answer']}"
    # we can either tokenize to IDs or to tokens; IDs is what the model actually uses:
    return len(tokenizer(prompt).input_ids)

# 5. Apply to each row
df["token_count"] = df.apply(count_prompt_tokens, axis=1)

# 6. (Optional) How many tokens remaining until the 2048 limit?
df["tokens_remaining"] = 2048 - df["token_count"]

# 7. Save back out
df.to_excel("/content/output_with_token_counts.xlsx", index=False)
print("Saved with columns 'token_count' and 'tokens_remaining'")
```
